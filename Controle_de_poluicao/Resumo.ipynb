- 1997 – Política Nacional de Recursos Hídricos.
- 1998 – Lei de Crimes Ambientais.
- 2000 – Sistema Nacional de Unidades de Conservação.

<h1 align="center"> Definições

<h1> # Definições importantes

Abaixo temos algumas definições importantes:

- **Poluição**: Qualquer contribuição natural ou artificial que provoque alteração temporária ou permanente no meio ambiente, que pode ser um ecossistema natural ou agrário, um sistema urbano ou até mesmo em microescala. (pode ou não haver riscos a saúde)
- **Contaminação**: Transmissão de substâncias ou microrganismos nocivos à saúde pela água. (pode ou não haver desequilibrio ambiental)
- **Poluente**: Toda e qualquer forma de matéria ou energia lançada nas águas, ar ou solo em quantidade, intensidade ou concentração em desacordo com os padrões de emissão estabelecidos por regulamento.
- **Fontes de poluição**: Toda e qualquer atividade, processo, operação, dispositivos móvel ou fixo que produzem ou podem produzir poluição ao meio ambiente.
- **Avaliação do impacto ambiental**: É um estudo visando identificar e prever o impacto ambiental no ambiente geofisico, saúde humana e/ou bem estar social. Bem como interpretar e comunicar os resultados a população. 
- **Corpo receptor**: Corpo hídrico superficial que recebe o lançamento de um efluente.
- **EIA**: Estudo de Impacto Ambiental.
- **RIMA**: Relatório de Impacto Ambiental.
- **IBAMA**: Instituto Brasileiro do Meio Ambiente e dos Recursos Naturais.
- **Água doce**: Presença de sais minerais em volta de 0.5 g/L.
- **Água salgada**: Presença de sais minerais acima de 3 g/L.
-  **Demanda Bioquímica de Oxigênio (DBO)**: Demanda de oxigênio dissolvido necessária para decomposição.
- **Demanda Química de oxigênio (DQO)**: Demanda de oxigênio necessária para oxidação da matéria orgânica, através de um agentequímico.


Ademais, temos os conceitos de projeto:

- **END OF PIPE**: Consiste no resíduo/despejo que sai da planta e é caracterizado e submetido ao tratamento adequado. (Tratamento do resíduo na saída)
- **IN PLANT**: Consiste na reformulação do processo de maneira a minimizar os custos de tratamento. (Diminuição do resíduo na planta)

Formas de poluição:

- Hidrica (das águas) - **lançamento** de liquidos.
- Atmosférica (do ar) - **emissão** de gases.
- Residuos sólidos (do solo) - **disposição** de sólidos.

<h1 align="center"> Poluição da água

<h1> # Poluição das águas

As poluições das águas podem ser dar de forma **pontual** ou **difusa**:

- **Cargas pontuais**: São introduzidas por lançamentos individualizados, como os que ocorrem no despejo de esgotos sanitários ou de efluentes industriais.
- **Cargas difusas**: São introduzidas por lançamentos passivos, como a lixiviação de agrotóxicos ou lixo de fazendas e cidades para o rio.


<h1> # Principais poluentes aquáticos

Abaixo temos os principais poluentes aquáticos:

- **Poluentes orgânicos biodegradáveis**: Degradado pelos organismos decompositores presentes no meio aquáticos.
- **Poluentes orgânicos recalcitrantes ou refratários**: Não são biodegradáveis ou sua taxa de biodegradação é muito lenta. (defensivos agrícolas, detergentes sintéticos, petróleo)
- **Metais**: Todos os metais podem ser solubilizados pela água, podendo gerar danos à saúde em função da quantidade ingerida.
- **Nutrientes**: O excesso de nutrientes gera o crescimento excessivo de alguns organismosaquáticos, acarretando prejuízo a determinados usos dos recursos hídricos superficiais e subterrâneos.
- **Organismos patogênicos**: Bactérias, vírus, protozoários, helmintos.
- **Sólidos em suspensão**: O aumento de sólidos aumenta a turbidez reduzindo as taxas de fotossíntese e prejudica a procura de alimento para algumas espécies, levando a desequilíbrios na cadeia alimentar.
- **Calor**: A temperatura da água afeta características físicas, químicas ebiológicas do meio aquático.

<h1> # Estágios de poluição das águas

A **poluição das águas** é dividida em alguns estágios, sendo eles:

- 1°: Poluição patogênica
    - Presença de microrganismos nocivos à saúde.
- 2°: Poluição total
    - Corpos receptores (CR) afetados pela carga poluidora que receberam, segundo a legislação.
- 3°: Poluição Química:
    - Maior diversidade de poluentes descarregada nos CR's.
    - Presença de substâncias não removidas pelo tratamentos convencional nos CR's.

<h1> # Padrões de qualidade da água

Os teores máximos de impurezas permitidos na água são estabelecidos em função dos seus usos. Desse modo, temos as seguintes classes:

- **Classe Especial**: Abastecimento para consumo humano com desinfecção.
- **Classe I**: Abastecimento para consumo humano após tratamento simplificado
- **Classe II**: Abastecimento para consumo humano após tratamento convencional.
- **Classe III**: Abastecimento para consumo humano após tratamento convencional ou avançado.
- **Classe IV**: Navegação e paisagística apenas.



Observação: Quanto maior a classe, mais impropria para consumo a água.

Abaixo temos os parametros para avaliação da qualidade da água:

- **Parâmetros físicos**:
    - Temperatura.
    - Sabor e odor.
    - Cor.
    - Turbidez.
    - Condutividade elétrica (ions dissolvidos).
    - Sólidos em suspensão.

    - Alcalinidade
    - Dureza (presença de Carbonato de calcio CaCO<sub>3</sub>)
    - Cloretos (sabor salgado)

    - Oxigênio dissolvido (OD)
    - Matéria orgânica e inorgânica.
    - Demanda Bioquímica de oxigênio
    - Demanda Química de oxigênio
- **Parâmetros Biológicos**:
    - Coliformes
    - Algas

<h1 align="center"> Autodepuração da água

<h1> # Mecanismos naturais que agem nos poluentes

Os poluentes, ao atingir os corpos de água, sofrem a ação de diversos mecanismos físicos, químicos e biológicos existentes na natureza.

Os principais **mecanismos físicos** estão descritos abaixo:

- **Diluição**: Diminuição da concentração do poluente pela sua difusão. 
- **Escoamento turbulento**: Aumenta a diluição, aumenta a penetração de oxigênio, aumenta a turbidez.
- **Gravidade** Realiza a sedimentação de particulas.
- **Luz**: Permite o crescimento da flora aquática e aumenta o oxigênio nas águas.
- **Temperatura**: Altera a solubilidade de gases e a cinética de algumas reações.

Os principais **mecanismos bioquímicos** estão descritos abaixo:

- **Retorno**: Seres decompositores que terminam a oxidação da matéria orgânica e completam o ciclo de nutrientes.
- **Decomposição aeróbica**: Decomposição com oxigênio no meio.
- **Decomposição anaeróbica**: Decomposição sem oxigênio no meio.

<h1> # Autodepuração

A autodepuração ocorre seguindo algumas etapas, sendo elas:

- Etapa 1: Decomposição
    - Nessa etapa é decomposto a matéria organica na água, é necessário satisfazer a **DBO**.
    - Quando os decompositores terminam sua tarefa dizemos que a matéria orgânica foi **estabilizada ou mineralizada**.
- Etapa 2: Recuperação do oxigênio dissolvido ou reaeração
    - Nessa etapa o oxigênio e readicionado as águas, seja pela atmosfera ou organismos vivos.

<h1 align="center"> Tratamento da água

<h1> # Tratamentos de despejos

O tratamento de despejos é dividida em alguns estágios, sendo eles:

- **Preliminar**: Remoção de sólidos grosseiros.
- **Tratamento primário**: Empregado para remoção de material visível em suspensão.
- **Tratamento Secundário**: Empregado para remoção, via ação biológica, do material em solução que é biodegradável.
- **Tratamento Terciário**: Visa a remoção do material que não foi removido nas etapas anteriores.



O tratamento dependerá do uso que água será empregada.

<h1> # Tratamento Preliminar

As etapas de **tratamento preliminar** consistem em:

- **Gradeamento**: Retirada de material grosseiro. (latas, madeiras, papelão)
- **Peneiramento**: Retirada de material menos grosseiro. (trapos, papéis, sacolas)
- **Desarenação**: Retirada de areias.
- **Neutralização**: Ajuste de pH. (pH = 7)
- **Equalização**: Homogenização do efluente.
