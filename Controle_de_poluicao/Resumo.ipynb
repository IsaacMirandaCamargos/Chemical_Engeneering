<h1 align="center"> História

<h1> # Pontos importantes

Abaixo temos os pontos importantes da legislação com relação ao meio ambiente:

- 1965 - Lei das florestas.
- 1967 - Lei da fauna silvestre, lei da pesca, código de minase mineração.
- 1972 - Declaração de Estocolmo sobre o ambiente humano.
- 1981 - Política Nacional do Meio Ambiente (O poluidor é obrigado a indenizar danos ambientaisque causar, independentemente da culpa) (Obrigatoriedade dos estudos e respectivosrelatórios de Impacto Ambiental, **EIA** e **RIMA**).
- 1989 – Criação do IBAMA.
- 1997 – Política Nacional de Recursos Hídricos.
- 1998 – Lei de Crimes Ambientais.
- 2000 – Sistema Nacional de Unidades de Conservação.

<h1 align="center"> Definições

<h1> # Definições importantes

Abaixo temos algumas definições importantes:

- **Poluição**: Qualquer contribuição natural ou artificial que provoque alteração temporária ou permanente no meio ambiente, que pode ser um ecossistema natural ou agrário, um sistema urbano ou até mesmo em microescala. (pode ou não haver riscos a saúde)
- **Contaminação**: Transmissão de substâncias ou microrganismos nocivos à saúde pela água. (pode ou não haver desequilibrio ambiental)
- **Poluente**: Toda e qualquer forma de matéria ou energia lançada nas águas, ar ou solo em quantidade, intensidade ou concentração em desacordo com os padrões de emissão estabelecidos por regulamento.
- **Fontes de poluição**: Toda e qualquer atividade, processo, operação, dispositivos móvel ou fixo que produzem ou podem produzir poluição ao meio ambiente.
- **Avaliação do impacto ambiental**: É um estudo visando identificar e prever o impacto ambiental no ambiente geofisico, saúde humana e/ou bem estar social. Bem como interpretar e comunicar os resultados a população. 
- **Corpo receptor**: Corpo hídrico superficial que recebe o lançamento de um efluente.
- **EIA**: Estudo de Impacto Ambiental.
- **RIMA**: Relatório de Impacto Ambiental.
- **IBAMA**: Instituto Brasileiro do Meio Ambiente e dos Recursos Naturais.
- **Água doce**: Presença de sais minerais em volta de 0.5 g/L.
- **Água salgada**: Presença de sais minerais acima de 3 g/L.

Ademais, temos os conceitos de projeto:

- **END OF PIPE**: Consiste no resíduo/despejo que sai da planta e é caracterizado e submetido ao tratamento adequado. (Tratamento do resíduo na saída)
- **IN PLANT**: Consiste na reformulação do processo de maneira a minimizar os custos de tratamento. (Diminuição do resíduo na planta)

Formas de poluição:

- Hidrica (das águas) - **lançamento** de liquidos.
- Atmosférica (do ar) - **emissão** de gases.
- Residuos sólidos (do solo) - **disposição** de sólidos.

<h1 align="center"> Poluição

<h1> # Estágios de poluição das águas

A **poluição das águas** é dividida em alguns estágios, sendo eles:

- 1°: Poluição patogênica
    - Presença de microrganismos nocivos à saúde.
- 2°: Poluição total
    - Corpos receptores (CR) afetados pela carga poluidora que receberam, segundo a legislação.
- 3°: Poluição Química:
    - Maior diversidade de poluentes descarregada nos CR's.
    - Presença de substâncias não removidas pelo tratamentos convencional nos CR's.

<h1> # Tratamento de efluente líquidos

O **tratamento de efluentes** é dividida em alguns etapas, sendo eles:

- 1° Fase
    - Presença de microrganismos nocivos à saúde.
- 2°: Poluição total

- 3°: Poluição Química:
    - Maior diversidade de poluentes descarregada nos CR's.
    - Presença de substâncias não removidas pelo tratamentos convencional nos CR's.
