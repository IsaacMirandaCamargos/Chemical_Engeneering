<h2> Inóculo:

Um inóculo é uma quantidade de micro-organismos, células ou tecidos que são introduzidos em um meio de cultura ou biorreator com o objetivo de iniciar ou propagar um processo de cultura ou fermentação. O inóculo atua como a "semente" que inicia o processo de crescimento e produção, que pode ser a fabricação de bioprodutos como bioetanol, proteínas terapêuticas, antibióticos, entre outros.

Ademais, iremos passar brevemente sobre cada etapa na produção de um inóculo:

- **Seleção e Preservação de Culturas de Micro-organismos**: A primeira etapa envolve a escolha da cultura de micro-organismos apropriada para o processo.
- **Preparação do Meio de Cultura de Inóculo**: O meio de cultura deve ser preparado de acordo com as necessidades nutricionais do micro-organismo (esterelizados).
- **Transferência da Cultura para o Meio de Inóculo**: A cultura de micro-organismos é então transferida para o meio de cultura preparado (esterelizados).
- **Incubação**: O inóculo é incubado para permitir que os micro-organismos se multipliquem.
- **Monitoramento do Crescimento**: O crescimento do inóculo é monitorado regularmente para garantir que os micro-organismos estejam crescendo de forma adequada.
- **Transferência do Inóculo para o Biorreator**: Uma vez que o inóculo atingiu a fase de crescimento desejada, é transferido para o biorreator onde o processo de fermentação ou cultura de células irá ocorrer.

<h2> Reator batelada:

O reator batelada é um dos principais tipos de biorreatores. Ele é chamado de batelada porque o processo de fermentação ou cultura celular ocorre em "lotes" ou ciclos. No início de cada batelada, o reator é preenchido com um meio de cultura estéril que contém todos os nutrientes necessários para o crescimento do micro-organismo. O inóculo é então adicionado e o reator é fechado.

A partir deste ponto, não há mais entrada ou saída de material, exceto para a possibilidade de adição de gases (como oxigênio ou dióxido de carbono). O micro-organismo cresce e consome os nutrientes do meio, possivelmente produzindo um produto de interesse. No final da batelada, o conteúdo do reator é drenado, o reator é limpo e esterilizado, e o processo começa novamente:



- **Equações de projeto**: As equações de projeto do reator batelada podem ser obtidas de diferentes formas, a partir do balanço do substrato apenas (reator ideal) e apartir do balanço do substrato e biomassa "X" (reator ideal, porem mais real).

<h2> Reator semi-Batelada:

Reatores semi-batelada, também conhecidos como reatores fed-batch, são uma variação dos reatores batelada em que a alimentação de substrato é adicionada ao reator durante o processo de fermentação ou cultura celular. Este tipo de reator é particularmente útil quando o crescimento dos micro-organismos é inibido por altas concentrações de substrato, pois permite a manutenção de uma concentração de substrato relativamente constante e baixa ao longo do tempo. 

Além disso, os reatores semi-batelada permitem um maior rendimento de biomassa ou produto por volume do reator, pois o volume do reator pode ser preenchido gradualmente ao longo do tempo, em vez de ser preenchido completamente no início:
