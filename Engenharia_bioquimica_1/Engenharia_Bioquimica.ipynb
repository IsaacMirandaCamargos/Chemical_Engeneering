<h1> # Lipdios

Os lipdios so compostos orgnicos insolveis e heterogneos em gua, porem solveis em solventes orgnicos.

Os lipdios so classificados em dois grandes grupos:

- **Simples**: So compostos que por hidrlise do origem somente a cidos graxos e lcool.
- **Complexos**: So compostos que apresentam outros grupos na molcula, alm dos cidos graxos e alcois.
