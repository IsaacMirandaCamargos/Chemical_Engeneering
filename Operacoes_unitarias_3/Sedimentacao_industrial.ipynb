<h1 align="center"> Introdução

<h1> # [ 1 ] Sedimentação

A sedimentação industrial consiste na utilização da força gravitacional para separar particulas de diferentes densidades.

Além disso, nesse processo podemos ter os seguintes **objetivos**:

- **Clarificação**: Fluido limpo é o objetivo.
- **Espessamento**: Sedimentar os sólidos para recuperá-los, seja para usar ou descartar.

Com relação as **vantagens do sendimentador industrial**, tem-se:

- Equipamento simples.
- Trabalha com altas capacidades (Q).
- Pode ser batelada ou contínuo.

<h1> # [ 2 ] Sedimentação: Fatores que influenciam

Abaixo temos alguns fatores que facilitam a sedimentação das particulas:

- Quanto menor o **tamanho**, mais facilmente sedimenta.
- Quanto maior a **esfericidade**, mais facilmente sedimenta.
- Quanto maior a **densidade**, mais facilmente sedimenta.
- Quando maior a **concentração**, mais dificilmente sedimenta.
