<h1 align="center"> Leito fluidizado

<h1> # Introdução:

O leito fluidizado é um processo amplamente utilizado na indústria química e petroquímica para realizar uma variedade de operações unitárias, como secagem, combustão, craqueamento, reforma e adsorção, entre outras.

Basicamente, o processo de leito fluidizado envolve a passagem de um fluido ascendente (normalmente um gás, como ar ou nitrogênio) através de um leito de partículas sólidas. A velocidade do fluido é controlada de forma que as partículas sejam suspensas no ar, formando um leito fluidizado. Neste estado, as partículas se comportam como um fluido e apresentam propriedades hidrodinâmicas únicas, como uma grande área superficial e alta taxa de transferência de massa e calor.

Além disso, tem-se as seguintes aplicações:

- Reações quimicas
- Secagem
- Recobrimento
- Granulação

<h1> # Vantangens e desvantagens:

Abaixo temos as vantagens do processo de leito fluidizado:

- Alta transferência de massa e calor
- Mistura homogênea e isotérmica
- Baixo custo de operação
- Permite operações em grandes escalas

Ademais, tem-se as seguintes desvantagens:

- Desgaste do equipamento
- Alguns materiais não fluidizam adequadamente

Onde:

- Fluidização coesiva: Fenômeno que ocorre quando as partículas sólidas em um leito fluidizado tendem a se aglomerar ou aderir umas às outras, dificultando o fluxo de fluido.

- Fluidização com aeração: Processo em que ar ou outro gás é forçado a passar através do leito de partículas sólidas, causando a suspensão das partículas.

- Fluidização do tipo areia: Processo em que as partículas sólidas são suspensas em um leito fluidizado e o fluxo de fluido é suficiente para manter as partículas em movimento e evitar a aglomeração.

- Fluidização do tipo jorro: Processo em que um gás é forçado através de um bico em uma câmara contendo partículas sólidas, causando a fluidização das partículas e criando uma corrente de jorro.

```python
class Leito_fluidizado:
    def __init__(self, dp, esferi, ps, pf=None, uf=None, Dc=None, Hml=None, M=None, unit="cm"):
        from math import pi
        self.dp = dp
        self.esferi = esferi
        self.ps = ps
        self.pf = pf
        self.uf = uf
        self.Dc = Dc
        self.Hml = Hml
        self.M = M
        self.A = pi*Dc**(2)/4
        self.g = 981 if unit == "cm" else 9.81
        self.poro_mf = 1 - M/(ps*self.A*Hml)
        self.B = 180 if esferi != 1 else 150
        
        if pf != None:
            self.dPmf = (ps - pf)*(1-self.poro_mf)*self.A*Hmf*self.g
            self.qmf =  (esferi)**(2) * (self.poro_mf)**(3) * (dp)**(2) * (ps - pf) * self.g
            self.qmf /= self.B * (1 - self.poro_mf) * self.uf
        else:
            self.dPmf = M*self.g/self.A
            self.qmf =  (esferi)**(2) * (self.poro_mf)**(3) * (dp)**(2) * (ps) * self.g
            self.qmf /= self.B * (1 - self.poro_mf) * self.uf
        
    def condicao_minima_de_fluidizacao(self):
        print("--------- RELATORIO ---------")
        print(f"Porosidade minima de fluidização = {self.poro_mf}")
        print(f"Queda de pressão minima de fluidização = {self.dPmf}")
        print(f"Velocidade minima de fluidização = {self.qmf}")


# Questão 17
leito = Leito_fluidizado(dp=52*10**-4,
                         esferi = 1,
                         ps = 2.6,
                         uf = 1.8*10**-4,
                         Dc = 3.2,
                         Hml = 26,
                         M = 267)
leito.condicao_minima_de_fluidizacao()

# Questão 18
leito = Leito_fluidizado(dp=0.009,
                         esferi = 0.8,
                         ps = 2.1,
                         uf = 2.3*10**-4,
                         Dc = 30,
                         Hml = 50,
                         M = 39000)
leito.condicao_minima_de_fluidizacao()

```

<h1 align="center"> Leito de Jorro

<h1> # Introdução:

No leito fluidizado comum, as partículas de sólidos são mantidas em suspensão em um fluxo contínuo de ar, com a ajuda de placas perfuradas ou outro dispositivo similar que mantém as partículas em um leito fluidizado. Já no leito de jorro, o ar é injetado em um ponto específico da câmara contendo as partículas de sólidos finos, criando um fluxo de ar ascendente que fluidiza os sólidos e forma um "leito" de partículas em suspensão.

Outra diferença importante entre os dois processos é a intensidade do fluxo de ar. No leito fluidizado comum, o fluxo de ar é relativamente baixo e constante, enquanto no leito de jorro o fluxo de ar é muito mais intenso e pulsante, criando um efeito de "jorro" que ajuda a fluidizar as partículas e promover a secagem e a granulação.
