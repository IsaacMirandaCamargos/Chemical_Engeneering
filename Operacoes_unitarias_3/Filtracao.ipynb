A filtração é a operação de seperação **sólido-fluido** que utiliza um meio poroso na retenção das partículas.

- **incompressivel**: Permeabilidade (k) e Porosidade ($\varepsilon$) **constantes**.
- **compressivel**: Permeabilidade (k) e Porosidade ($\varepsilon$) **variam com o tempo**.
