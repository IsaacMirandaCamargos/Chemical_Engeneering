<h1 align="center"> Filtro Prensa

<h1> # Introdução:

O filtro prensa consiste em uma série de placas e quadros que são colocados juntos, formando câmaras de filtração. Cada câmara é preenchida com uma camada de material filtrante, como tecido ou papel filtro, que permite que o líquido passe através dele, enquanto os sólidos são retidos.

O processo de filtração começa com a alimentação da mistura de sólidos e líquidos na câmara de filtragem. À medida que a pressão é aplicada, o líquido atravessa o material filtrante e flui para fora do filtro, deixando os sólidos retidos dentro da câmara de filtragem. Esses sólidos formam uma torta de filtração que pode ser removida do filtro prensa após o término do processo de filtração.
