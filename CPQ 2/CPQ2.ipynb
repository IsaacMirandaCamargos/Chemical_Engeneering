# <u> MALHA FECHADA:

Entendido como fazer a equação de transferência de um sistema baseado no diagrama de blocos, podemos determinar a equação de transferência de um sistema em malha fechada.



Rearranjando a equação, temos: (SISO)

- ### $Y(s) = \frac{K_mG_cG_vG_p}{1 + G_mG_cG_vG_p}Y_{ps}(s) + \frac{G_D}{1 + G_mG_cG_vG_p}D(s)$

Desse modo, podemos escrever a equação de transferência de um sistema para definir um **set point** ou um **regulador**:

- **set point**: (servo)
    - ### $\frac{Y(s)}{Y_{ps}(s)} = \frac{K_mG_cG_vG_p}{1 + G_mG_cG_vG_p}$

- **regulador**:
    - ### $\frac{Y(s)}{D(s)} = \frac{G_D}{1 + G_mG_cG_vG_p}$

Onde:

- $Y(s)$: é a variável de controle.
- $Y_{ps}(s)$: é a variável manipulada.
- $D(s)$: é a variável de distúrbio.
- $K_m$: Ganho do medidor
- $G_c$: FT de controle
- $G_v$: FT da válvula (elemento final)
- $G_p$: FT da planta
- $G_m$: FT do medidor
- $G_D$: FT do distúrbio



**Observação**: Para sistema MIMO a equação será um sistema de equações, matriz, ou seja é impossivel passar dividindo. Nesse caso aplicamos a inversa da matriz para isolar a variável controlada.

# <u> MALHA FECHADA: Equação geral

Ademais, podemos escrever como montar a equação de transferência de um sistema em malha fechada de modo geral:

- ## $\frac{Z}{Z_i} = \frac{\prod_{f}}{1 + \prod_{e}}$

Onde:

- $Z$: é a variável de saída.
- $Z_i$: é a variável de entrada.
- $\prod_{f}$: é o produto dos ganhos da entrada ($U_i$) até a variável de saída ($Y$).
- $\prod_{e}$: é o produto dos ganhos do reciclo até a variável de saída ($Y$)

# <u> PID:

Um controlador PID (Proporcional–Integral–Derivativo) ajusta sua saída com base em três ações sobre o erro (diferença entre valor desejado e valor medido):

* **Proporcional (P):** reage instantaneamente ao erro, gerando correção proporcional à sua magnitude.
* **Integral (I):** acumula erros passados para eliminar desvios persistentes (offset).
* **Derivativo (D):** antecipa a tendência do erro, agindo contra variações rápidas.

Combinando P, I e D, o PID fornece controle estável, rápido e com mínimo overshoot.

O controlador PID é expresso matematicamente como:

- **Paralela**:
    - #### $\boxed{G_c = \frac{P(s)}{E(s)} = K_c\left(1 + \frac{1}{\tau_i s} + \tau_d s\right)}$

- **Série**:
    - #### $\boxed{G_c = \frac{P(s)}{E(s)} = K_c\left(1 + \frac{1}{\tau_i s + 1}\right) \left(1 + \tau_d s\right)}$

**Observação**: As deduções estão a seguir.

# <u> PID: Proporcional

Dito isto, o controlador proporcional pode ser representado como:

- ### $p(t) = p(t-1) + K_p\bigl(y_{sp}(t) - y(t)\bigr)$

E a função de transferência do controlador proporcional é dada por:

- ##### $p(t) - p(t-1) = K_p \bigl(y_{sp}(t) - y(t)\bigr)$

- ##### $\mathcal{L}\{p(t) - p(t-1)\} = K_p \,\mathcal{L}\{y_{sp}(t) - y(t)\}$

- ##### $P(s) = K_p\,Y(s)$

- ### $\boxed{G(s) = \frac{P(s)}{Y(s)} = K_p}$


Onde:
- $p(t)$: é a variável de controle.
- $p(t-1)$: é a variável de controle do instante anterior.
- $K_p$: é o ganho proporcional.
- $y_{sp}(t)$: é o set point.
- $y(t)$: é a variável de saída.

# <u> PID: Integral

O controlador integral pode ser representado como:

- ##### $e(t) = y_{sp}(t) - y(t)$

- ### $p(t) - p(t-1) = \frac{1}{\tau_{I}}\displaystyle\int_{0}^{t} e(t)\mathrm{d}t$

Já a função de transferência do controlador integral é dada por:

- ##### $\mathcal{L}\{p(t) - p(t-1)\} = \frac{1}{\tau_{I}} \mathcal{L}\{\displaystyle\int_{0}^{t} e(t)\mathrm{d}t\}$

- ##### $P(s) = \frac{1}{\tau_{I}s}E(s)$

- ### $\boxed{G(s) = \frac{P(s)}{Y(s)} = \frac{1}{\tau_{I}s}}$


# <u> PID: Derivativo

O controlador derivativo pode ser representado como:

- ##### $e(t) = y_{sp}(t) - y(t)$

- ### $p(t) - p(t-1) = \tau_{D}\frac{\mathrm{d}e}{\mathrm{d}t}$

Aplicando a transformada de Laplace, temos:

- ##### $\mathcal{L}\{p(t) - p(t-1)\} = \tau_{D} \mathcal{L}\{\frac{\mathrm{d}e}{\mathrm{d}t}\}$

- ##### $P(s) = \tau_{D}sE(s)$

- ### $\boxed{G(s) = \frac{P(s)}{Y(s)} = \tau_{D}s}$

- #### $\boxed{\frac{Y(s)}{Y_{sp}(s)} = \frac{G_c G}{1 + G_c G}}$

# <u> ESTABILIDADE:

Um sistema de controle feedback é estável se, para toda entrada limitada, sua saída também for limitada. Desse modo, para um sistema ser estável, é necessário que o denominador da FT tenha todas as raizes reais negativas.

Ademais, a função de transferência pode ser expressa como:

$$
\frac{Y(s)}{Y_{ep}(s)} = K \frac{(s - z_1)\cdots(s - z_m)}{(s - p_1)\cdots(s - p_n)}
$$

* **Polos com parte real negativa**: sistema estável
* **Polos com parte real positiva**: sistema instável
* **Polos no eixo imaginário**: sistema marginalmente estável ($Re < 0$) ou instável ($Re > 0$), dependendo da multiplicidade

| Tipo de Pólo                | Resposta Temporal                        |
| --------------------------- | ---------------------------------------- |
| Real negativo               | Exponencial decrescente (sem oscilações) |
| Complexo com parte negativa | Oscilação amortecida                     |
| Real positivo               | Crescimento exponencial (instável)       |
| Complexo com parte positiva | Oscilação crescente (instável)           |



# <u> ESTABILIDADE: Método de Routh

Considerando a função de transferência:

$$
a_n s^n + a_{n-1} s^{n-1} + \cdots + a_1 s + a_0 = 0
$$

Podemos dizer que ela é instavel se:

- Pelo menos 1 coeficiente for negativo.
- Caso todos os coeficientes forem positivos, mas a primeira coluna da tabela de Routh tiver um elemento negativo.

Desse modo, podemos montar a tabela de Routh da seguinte forma:

| Ordem | Coluna 1      | Coluna 2      |
|-------|---------------|---------------|
| 1     | $a_n$         | $a_{n-2}$     | 
| 2     | $a_{n-1}$     | $a_{n-3}$     |
| 3     | $b_1$         | $b_2$         |
| 4     | $c_1$         | $c_2$         |
| ...   | ...           | ...           |
| n + 1 | $z_1$         | $z_2$         |

Onde:

- ### $b_1 = \frac{a_{n-1} a_{n-2} - a_{n} a_{n-3}}{a_{n-1}}$
- ### $b_2 = \frac{a_{n-1} a_{n-4} - a_{n} a_{n-5}}{a_{n-1}}$

- ### $c_1 = \frac{b_1 a_{n-3} - b_2 a_{n-1}}{b_1}$
- ### $c_2 = \frac{b_1 a_{n-5} - b_2 a_{n-3}}{b_1}$

```python
#----------------------------------------------------------------------------//
# BIBLIOTECAS
#----------------------------------------------------------------------------//
import numpy as np
import matplotlib.pyplot as plt
import control as ct

from rich.console import Console
from rich.table import Table
from rich.rule import Rule
from rich.panel import Panel
from rich import box

#----------------------------------------------------------------------------//
# CONFIGURAÇÃO DO CONSOLE RICH
#----------------------------------------------------------------------------//
console = Console()

def print_header(title: str):
    console.print(Rule(title, style="bold bright_magenta"))

def print_transfer_functions(Gp, Gm, Gc, FT):
    table = Table(
        title="📊 Funções de Transferência",
        box=box.DOUBLE_EDGE,
        show_header=True,
        header_style="bold bright_cyan",
        border_style="bright_magenta"
    )
    table.add_column("Elemento", style="bold bright_white", justify="center")
    table.add_column("FT", style="bright_white", overflow="fold")

    table.add_row("Planta (Gp)", f"[bright_yellow]{Gp}[/]")
    table.add_row("Medidor (Gm)", f"[bright_green]{Gm}[/]")
    table.add_row("Controlador (Gc)", f"[bright_blue]{Gc}[/]")
    table.add_row("Malha Fechada (FT)", f"[bright_red]{FT}[/]")

    console.print(table)

#----------------------------------------------------------------------------//
# PROGRAMA PRINCIPAL
#----------------------------------------------------------------------------//
if __name__ == "__main__":
    # Entradas e variáveis
    Tts = 40
    Tsp = 10
    s = ct.tf('s')
    t = np.arange(0, 301, 1)

    # Definição das FTs
    Gp = 0.018898 / (10*s**3 + 5.1135*s**2 + 0.82503*s + 0.041368)
    Gm = (-3.7066435*s + 0.53333) / (13.9*s**2 + 8.95*s + 1)
    Gc = (1008.45*s**2 + 134.46*s + 4.482) / (30*s)
    FT = (Gc * Gp) / (1 + Gc * Gp * Gm)

    # Exibe as tabelas no notebook
    print_header("💡 Transfer Functions")
    print_transfer_functions(Gp, Gm, Gc, FT)

    # Sinal de entrada e simulação
    u1 = (Tsp - Tsp) * np.ones_like(t[:1])
    u2 = (Tsp - Tts) * np.ones_like(t[1:])
    u_m = np.concatenate((u1, u2))
    _, ym = ct.forced_response(Gm, T=t, U=u_m)
    _, y  = ct.forced_response(FT, T=t, U=ym)

    # Ajuste de offset
    y   += Tts
    u_m += Tts
    ym  += Tts

    #----------------------------------------------------------------------------//
    # GRÁFICOS
    #----------------------------------------------------------------------------//
    print_header("📈 Resultados da Simulação")
    plt.figure(figsize=(10, 5))
    plt.plot(t, u_m, linestyle='--', label='Setpoint (°C)')
    plt.plot(t, y,      linestyle='-',  label='Saída (°C)')
    plt.plot(t, ym,     linestyle=':',  label='Medida (°C)')
    plt.title("Resposta da Malha de Controle", pad=15)
    plt.xlabel("Tempo (s)")
    plt.ylabel("Temperatura (°C)")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()
```
