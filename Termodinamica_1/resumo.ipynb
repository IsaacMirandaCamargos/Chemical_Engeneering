<h1 align="center"> Definições e conceitos

<h2><u> Sistemas Termodinâmicos:

Um sistema termodinâmico é uma porção do universo que está sendo estudada e é definida pelo observador. Tudo o que está fora do sistema é chamado de vizinhança. A fronteira é o limite que separa o sistema da vizinhança. 

A natureza dessa fronteira (seja ela fixa, móvel, permeável ao calor, impermeável à matéria, etc.) determina a maneira como o sistema pode interagir com a vizinhança. Desse modo, é possivel classificar os sistemas entre:

- **Sistema aberto**: é um sistema que pode trocar tanto energia (na forma de calor ou trabalho) quanto matéria com a vizinhança.
- **Sistema fechado**: é um sistema que pode trocar energia, mas não matéria, com a vizinhança.
- **Sistema isolado**: é um sistema que não troca nem energia nem matéria com a vizinhança.

<h2>  Estado de um sistema:

O estado de um sistema termodinâmico é determinado pelos valores das propriedades macroscópicas:

- **Pressão**
- **Temperatura**
- **Volume**
- **Composição química** (para misturas)

O estado de um sistema independe do caminho percorrido para atingir esses valores.

<h2> Estado de um Sistema: Classificação

Ademais, os estados termodinâmicos podem ser classificados mediante sua variação em relação ao tempo:

- **Estado Estacionário ou Permanente:** Os valores de todas as variáveis do processo não variam com o tempo.
- **Estado Transiente ou Não-Estacionário:** Pelo menos uma variável do processo varia com o tempo.

**Observação**: Um sistema em **equilíbrio** é um sistema em que todas as variáveis termodinâmicas não possuem gradientes em relação a posição.

<h2> Classificação das propriedades:

Dentro do estudo da termodinâmica, existem duas categorias importantes de propriedades: propriedades intensivas e propriedades extensivas:

- **Propriedades intensivas**: Não dependem da quantidade de substância presente no sistema.
- **Propriedades extensivas**: Dependem da quantidade de substância presente no sistema. (ex: Entalpia)


<h2> Regra das fases:

A Regra das Fases fornece o número mínimo de variáveis intensivas que devem ser especificadas para determinar completamente o estado de um sistema. Por exemplo, para um sistema de uma única substância, a regra das fases afirma que precisamos especificar duas variáveis intensivas (como pressão e temperatura) para determinar completamente o estado do sistema.

- **Regra das fases**: Onde (F: gráus de liberadade; C: componentes da mistura; P: Número de fases da mistura)
 
<h3 align="center"> $F = C - P + 2$


<h2><u> Lei zero da Termodinâmica:

A Lei Zero da Termodinâmica afirma que:

- **Se dois sistemas, A e B, estão em equilíbrio térmico com um terceiro sistema, C, então A e B estão em equilíbrio térmico um com o outro**.

**Observação**: Essa lei é fundamental para a termodinâmica, pois permite a definição de uma escala de temperatura e o conceito de igualdade de temperatura.

<h2><u> Trabalho (W):

Trabalho em termodinâmica é definido como a transferência de energia que ocorre em virtude de uma força que provoca um deslocamento.

- **Equações de trabalho**: O trabalho é geralmente representado pela letra (W) e pode ser calculado pelo produto da força (F) pelo deslocamento (d).

<h3 align="center">
    1°: $W = F \cdot d$<br>
    2°: $W = -P .\Delta V$
</h3>

- **Convenção**: A convenção de sinais para o trabalho nesta disciplina será:

    - $W > 0$: trabalho realizado sobre o sistema (pela vizinhança)
    - $W < 0$: trabalho realizado pelo sistema (sobre a vizinhança)

<h2><u> Calor:

O calor é uma forma de energia que é transferida entre sistemas devido a diferenças de temperatura. O calor flui do sistema de maior temperatura para o de menor temperatura até que ambos atinjam um equilíbrio térmico.

A quantidade de calor transferido é denotada por $Q$ e, assim como o trabalho, não é uma propriedade do sistema, mas sim uma quantidade que depende do caminho percorrido durante o processo termodinâmico.

- **Convenção**: A convenção de sinais para o calor será:

    - $Q > 0$: calor transferido para o sistema (absorvido pelo sistema)
    - $Q < 0$: calor transferido do sistema (liberado pelo sistema)


<h2><u> Formas de energia:

A energia pode existir de muitas formas diferentes, e entender essas diferentes formas de energia é essencial para a física e a engenharia, incluindo a termodinâmica.

- **Energia cinética** (KE): é a energia associada ao movimento de um objeto.(v: velocidade; m: massa)
<h3 align="center">
    $EK = \frac{1}{2}mv^2$
</h3>
- **Energia potencial gravitacional** (PE): é a energia associada à posição de um objeto em um campo gravitacional. (m: massa; g: aceleração grav.; h: altura)
<h3 align="center">
    $EP = mgh$
</h3>
- **Energia potencial elástica** (EE): a energia armazenada em um objeto quando ele é esticado ou comprimido. (k: constante da mola; x: deslocamento)
<h3 align="center">
    $EE = \frac{1}{2}kx^2$
</h3>
- **Energia térmica** (Q): a energia interna de um sistema devido ao movimento aleatório de suas partículas. (m: massa; C: capacidade calorifica; T: Temperatura)
<h3 align="center">
    $Q = mC\Delta T$
</h3>
- **Energia elétrica** (P): a energia que é transportada por uma corrente elétrica. 
<h3 align="center">
    $P = IV$
</h3>

Ademais, define-se o **Principio da Conservação da Energia Mecânica** como sendo:

<h3 align="center">
    $ \Delta EK + \Delta EP = 0$
</h3>

<h2><u> Primeira Lei da termodinâmica:

A primeira lei da termodinâmica é uma extensão da lei de conservação de energia, aplicada para sistemas termodinâmicos:

- **A energia não pode ser criada nem destruída, mas apenas transferida ou transformada.**

Desse modo, chegamos na expressão:



<h2><u> Energia interna: (U)

A energia interna, no contexto termodinâmico, representa a totalidade da energia conservada em um dado sistema. Essa energia é a agregação de diversas formas de energia microscópicas que são atribuídas às partículas - como átomos, moléculas e íons - que formam o sistema. 

Ademais, é impossivel calcular a energia interna (U) diretamente, sendo possivel apenas o calculo de sua variação a partir de um estado de referência.

- **Equacões de energia interna**: A energia interna pode ser representada em função do número de mols (n) de uma substância ou em função de um sistema como um todo (U: J ou J/mol)

<h3 align="center"> 1°: $\Delta U = Q - W$
<h3 align="center"> 2°: $dU = dQ - dW$ 
<h3 align="center"> 2°: $d(nU) = dQ - dW$ 

<h2><u> Entalpia: (H)

Na termodinâmica, a entalpia é uma propriedade crucial que representa a energia total contida em um sistema. Incorpora a energia interna, que engloba a energia requerida para estabelecer o sistema, assim como a energia demandada para propiciar o deslocamento do ambiente para acomodar o sistema. 

Em contextos onde a pressão é mantida constante, a variação na entalpia do sistema é diretamente equivalente à quantidade de calor que foi absorvida pelo sistema.

- **Equacões de entalpia**: A entalpia pode ser representada em função do número de mols (n) de uma substância ou em função de um sistema como um todo (U: J ou J/mol)

<h3 align="center"> 1°: $\Delta H = \Delta U + \Delta PV$
<h3 align="center"> 2°: $dH = dU + d(PV)$ 
<h3 align="center"> 2°: $d(nH) = d(nU) + d(nPV)$ 

<h2><u> Variações de estados:

Na termodinâmica, um processo é uma mudança de estado de um sistema de um estado inicial para um estado final. Existem vários tipos de processos, que podem ser classificados com base em diferentes critérios, como a transferência de calor, a mudança de pressão, a mudança de volume, entre outros. 

Além disso, um critério importante para a classificação de processos é a sua reversibilidade ou irreversibilidade:

- **Processos reversiveis**: Um processo reversível é aquele que pode ser invertido de tal maneira que tanto o sistema quanto a sua vizinhança retornem aos seus estados iniciais sem qualquer traço de alterações. Durante tal processo, o sistema permanece em equilíbrio termodinâmico em todas as etapas e nenhuma energia é dissipada. Os processos reversíveis são ideais e raramente, se é que existem, ocorrem na realidade.

- **Processos irreversiveis**: Um processo irreversível é aquele que não pode ser revertido sem deixar alterações no sistema ou na sua vizinhança. Em outras palavras, o sistema e a vizinhança não podem ser retornados aos seus estados iniciais apenas invertendo o processo. A maioria dos processos naturais são irreversíveis, ocorrendo de maneira espontânea em uma única direção. Isso inclui processos que envolvem dissipação de energia, como fricção e transferência de calor através de uma diferença finita de temperatura. 



<h2> Processos:

Ademais, teremos diversos processos comuns na termodinâmica e avaliaremos ele a partir da variação de energia interna.

- **Processo a volume constante**: Se volume constante temos (dW = 0)

<h3 align="center"> 1°:$dU = dQ$</h3>
<h3 align="center"> 2°:$d(nU) = dQ$</h3>
    
- **Processo a pressão constante**: Se pressão constante temos
    
    - **Dedução**: ($H = U + PV$)
        - $dU = dQ + dW$
        - $dU + dPV = dQ$
        - $d(U + PV) = dQ$; 
        - $dH = dQ$
        
    
<h3 align="center"> 1°:$dH = dQ$</h3>
<h3 align="center"> 1°:$d(nH) = dQ$</h3>

<h2> Capacidade calorifica:

A capacidade calorífica é uma característica fundamental de uma substância que quantifica a quantidade de calor requerida para elevar a temperatura da substância em uma quantidade específica. Essencialmente, ela fornece uma medida da eficiência com que uma substância pode reter energia térmica.

- **Capacidade calorifica a volume constante**:
    
<h3 align="center"> 1°:$C_v = (\frac{dU}{dT})_v$</h3>

- **Capacidade calorifica a pressão constante**:
    
<h3 align="center"> 1°:$C_p = (\frac{dH}{dT})_P$</h3>
