<h1 align="center"> Definições e conceitos (1)

<h2> Sistemas Termodinâmicos:

Um sistema termodinâmico é uma porção do universo que está sendo estudada e é definida pelo observador. Tudo o que está fora do sistema é chamado de vizinhança. A fronteira é o limite que separa o sistema da vizinhança. 

A natureza dessa fronteira (seja ela fixa, móvel, permeável ao calor, impermeável à matéria, etc.) determina a maneira como o sistema pode interagir com a vizinhança. Desse modo, é possivel classificar os sistemas entre:

- **Sistema aberto**: é um sistema que pode trocar tanto energia (na forma de calor ou trabalho) quanto matéria com a vizinhança.
- **Sistema fechado**: é um sistema que pode trocar energia, mas não matéria, com a vizinhança.
- **Sistema isolado**: é um sistema que não troca nem energia nem matéria com a vizinhança.

<h2> Classificação das propriedades:

Dentro do estudo da termodinâmica, existem duas categorias importantes de propriedades: propriedades intensivas e propriedades extensivas:

- **Propriedades intensivas**: Não dependem da quantidade de substância presente no sistema.
- **Propriedades extensivas**: Dependem da quantidade de substância presente no sistema. (ex: Entalpia)


<h2>  Estado de um sistema:

O estado de um sistema termodinâmico é determinado pelos valores das propriedades macroscópicas:

- **Pressão**
- **Temperatura**
- **Volume**
- **Composição química** (para misturas)

O estado de um sistema independe do caminho percorrido para atingir esses valores.

<h2> Estado de um Sistema: Classificação

Ademais, os estados termodinâmicos podem ser classificados mediante sua variação em relação ao tempo:

- **Estado Estacionário ou Permanente:** Os valores de todas as variáveis do processo não variam com o tempo.
- **Estado Transiente ou Não-Estacionário:** Pelo menos uma variável do processo varia com o tempo.

**Observação**: Um sistema em **equilíbrio** é um sistema em que todas as variáveis termodinâmicas não possuem gradientes em relação a posição.

<h1 align="center"> Definições e conceitos (2)

<h2> Lei zero da Termodinâmica:

A Lei Zero da Termodinâmica afirma que:

- **Se dois sistemas, A e B, estão em equilíbrio térmico com um terceiro sistema, C, então A e B estão em equilíbrio térmico um com o outro**.

**Observação**: Essa lei é fundamental para a termodinâmica, pois permite a definição de uma escala de temperatura e o conceito de igualdade de temperatura.

<h2> Trabalho (W):

Trabalho em termodinâmica é definido como a transferência de energia que ocorre em virtude de uma força que provoca um deslocamento.

- **Equações de trabalho**: O trabalho é geralmente representado pela letra (W) e pode ser calculado pelo produto da força (F) pelo deslocamento (d).

<h3 align="center">
    1°: $W = F \cdot d$<br>
    2°: $W = -P .\Delta V$
</h3>

- **Convenção**: A convenção de sinais para o trabalho nesta disciplina será:

- $W > 0$: trabalho realizado sobre o sistema (pela vizinhança)
- $W < 0$: trabalho realizado pelo sistema (sobre a vizinhança)

<h2> Calor:

O calor é uma forma de energia que é transferida entre sistemas devido a diferenças de temperatura. O calor flui do sistema de maior temperatura para o de menor temperatura até que ambos atinjam um equilíbrio térmico.

A quantidade de calor transferido é denotada por $Q$ e, assim como o trabalho, não é uma propriedade do sistema, mas sim uma quantidade que depende do caminho percorrido durante o processo termodinâmico.

- **Convenção**: A convenção de sinais para o calor será:

- $Q > 0$: calor transferido para o sistema (absorvido pelo sistema)
- $Q < 0$: calor transferido do sistema (liberado pelo sistema)


<h2> Formas de energia:

A energia pode existir de muitas formas diferentes, e entender essas diferentes formas de energia é essencial para a física e a engenharia, incluindo a termodinâmica.

- **Energia cinética** (KE): é a energia associada ao movimento de um objeto.(v: velocidade; m: massa)


<h3 align="center">
    $EK = \frac{1}{2}mv^2$
</h3>

- **Energia potencial gravitacional** (PE): é a energia associada à posição de um objeto em um campo gravitacional. (m: massa; g: aceleração grav.; h: altura)

<h3 align="center">
    $EP = mgh$
</h3>


- **Energia potencial elástica** (EE): a energia armazenada em um objeto quando ele é esticado ou comprimido. (k: constante da mola; x: deslocamento)

<h3 align="center">
    $EE = \frac{1}{2}kx^2$
</h3>


- **Energia térmica** (Q): a energia interna de um sistema devido ao movimento aleatório de suas partículas. (m: massa; C: capacidade calorifica; T: Temperatura)

<h3 align="center">
    $Q = mC\Delta T$
</h3>

- **Energia elétrica** (P): a energia que é transportada por uma corrente elétrica. 

<h3 align="center">
    $P = IV$
</h3>

Ademais, define-se o **Principio da Conservação da Energia Mecânica** como sendo:

<h3 align="center">
    $ \Delta EK + \Delta EP = 0$
</h3>
