```python
import matplotlib.pyplot as plt
import numpy as np
from math import log, pi

Kiso = 1.4  # W/mK
D = 10 * 10**(-3)
r1 = D / 2
T1 = 200  # C
Tinf = 25  # C
h = 140  # W/m²C

# Calculate delta_T
delta_T = Tinf - T1

# Generate r_iso values
r_iso = np.linspace(r1, r1 * 10, 1000)
q_L_values = []  # Lista para armazenar os valores calculados de q_L

for r in r_iso:
    # Calculate Re
    Re = (1 / (2 * pi * r * h)) + (log(r / r1) / (2 * pi * Kiso))
    # Calculate q_L and store it
    q_L = delta_T / -Re

    q_L_values.append(q_L)

# Plotting
plt.plot(r_iso, q_L_values)  
plt.xlabel('r_iso [m]')
plt.ylabel('q_L [W/m]')
plt.title('q_L em função de r_iso')
plt.grid(True)
plt.show()
```

Calcule a resistência térmica total, por unidade de comprimento do tubo, em um tubo de cobre com diâmetro interno de 8 mm e diâmetro externo de 10 mm, possuindo as seguintes espessuras de isolamento: 0 – 2 – 5 – 10 – 20 – 40 mm
O isolamento é composto por vidro celular e o coeficiente de transferência de calor por convecção em sua superfície externa é de 5 W/m².K

Dados:
- K_COBRE = 401 W/m.K
- K_VIDRO = 0,055 W/m.K



```python
from math import log, pi

# Constantes e medidas dadas
Di = 8 * 10**(-3) # Diâmetro interno em metros
De = 10 * 10**(-3) # Diâmetro externo em metros
r1 = Di / 2 # Raio interno em metros
r2 = De / 2 # Raio externo em metros
h = 5 # Coeficiente de transferência de calor por convecção externa em W/m²K
K_cobre = 401 # Condutividade térmica do cobre em W/mK
K_vidro = 0.055 # Condutividade térmica do vidro em W/mK

# Espessuras de isolamento dadas em milímetros, convertidas para metros
espessuras_mm = [0, 2, 5, 10, 20, 40]
espessuras_m = [e * 10**(-3) for e in espessuras_mm] # Conversão para metros

# Lista para armazenar a resistência térmica total para cada espessura
resistencias_totais = []

# Cálculo da resistência térmica para cada espessura
for esp in espessuras_m:
    r_externo_total = r2 + esp # Raio externo total com isolamento
    R_conducao_cobre = log(r2 / r1) / (2 * pi * K_cobre) # Resistência do cobre
    R_conducao_vidro = log(r_externo_total / r2) / (2 * pi * K_vidro) if esp != 0 else 0 # Resistência do vidro, se houver isolamento
    R_conveccao_externa = 1 / (h * 2 * pi * r_externo_total) # Resistência por convecção externa
    R_total = R_conducao_cobre + R_conducao_vidro + R_conveccao_externa
    resistencias_totais.append((esp, R_total))

resistencias_totais
```
