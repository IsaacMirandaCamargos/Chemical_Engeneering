```python
# Bibliotecas
import numpy as np
from numpy import log as ln
import math
```

```python
# Lei de Rittinger (n=2): Moagem Fina
C = 6.4 # t/h
k = 0.134 #Adimensional
D1 = 0.3 #cm
D2 = 0.1 #cm

W = C*k*((1/D2)-(1/D1))

print(f"O consumo foi (-W): {W:.1f}")
```

```python
# Lei de Kick (n=1): aplicada às primeiras fases do britamento

C = 100 # t/h
k = 0.134 #Adimensional
D1 = 5 #cm
D2 = 0.236 #cm

W = C*k*ln(D1/D2)

print(f"O consumo foi (-W): {W:.1f}")
```

```python
# Lei de Bond (n=1,5): lei geral
C = 100 # t/h
k = 0.134 #Adimensional
Wi = 12.74 # kWh/t (A ÚMIDO)
Wi = Wi * 1.34 # kWh/t (A SECO)
D1 = 5 #cm
D2 = 0.236 #cm

W = C*k*Wi*((1/np.sqrt(D2))-(1/np.sqrt(D1)))

print(f"O consumo foi (-W): {W:.1f}")
```

```python
# Gasto de energia se reduzir espaçamento pela metade

# Dados
W_vazio=1/4
W_primeiro = 14
Wo = W_primeiro - W_vazio # Potência do material britado
D1o = 0.3 #cm
D2o = 0.1 #cm
D2f = D2o/2 #cm reduziu pela metade

# Lei de Bond
W1 = (((1/np.sqrt(D2f))-(1/np.sqrt(D1o)))/((1/np.sqrt(D2o))-(1/np.sqrt(D1o)))) * Wo

# Lei de Kick
W2 = (ln(D1o/D2f)/ln(D1o/D2o)) * Wo

# Lei de Rittinger
W3 = (((1/D2f)-(1/D1o))/((1/D2o)-(1/D1o))) * Wo

print(f"Lei de Bond - O consumo foi (-W): {W1:.1f}")
print(f"Lei de Kick - O consumo foi (-W): {W2:.1f}")
print(f"Lei de Rittinger - O consumo foi (-W): {W3:.1f}")
```
