# Operações Unitárias 4: Resumo

As operações unitárias são os processos fundamentais aplicados em engenharia química, utilizados para modificar propriedades físicas ou químicas dos materiais, visando à produção de bens de consumo ou insumos industriais. A disciplina "Operações Unitárias 4" foca em alguns processos específicos, como evaporação, secagem, cristalização e extração.

- **Evaporação**: Processo no qual a água ou outro solvente é removido de uma solução por meio do aquecimento, gerando uma solução mais concentrada. Este processo é amplamente utilizado em indústrias alimentícias e químicas.

- **Secagem**: Visa a remoção de umidade de materiais sólidos, com o uso de calor ou ar quente, transformando um material úmido em um produto seco. É um processo comum em setores como alimentício e farmacêutico.

- **Cristalização**: Utilizada para separar componentes de uma solução por meio da formação de cristais sólidos. É uma técnica essencial para purificação de substâncias e obtenção de produtos de alto valor agregado.

- **Extração**: Processo de separação onde se usa um solvente adequado para isolar um determinado componente de uma mistura. Este processo pode ser líquido-líquido ou sólido-líquido, sendo aplicado em várias indústrias, como a de óleo e a farmacêutica.

Essas operações desempenham papel fundamental no desenvolvimento de processos industriais eficientes, permitindo a transformação de matérias-primas em produtos com as características desejadas.



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Definindo a função que representa a EDO
def func(y, t):
    return y
# Condição inicial
y0 = 1
# Intervalo de integração
t = np.linspace(0, 2, 100)
# Resolvendo a EDO
sol = odeint(func, y0, t)
# Plotando a solução
plt.plot(t, sol, label='Solução Numérica (odeint)')
plt.xlabel('t')
plt.ylabel('y')
plt.title("Solução da EDO $y' = y$ usando odeint")
plt.legend()
plt.show()
```
