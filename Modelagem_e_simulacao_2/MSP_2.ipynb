<h1> # Linearização de modelos não lineares

A linearização de modelos não lineareas consiste em:

- Encontrar a aproximação linear de uma função em dado ponto **(hiperplano tangente)**. 
- No estudo de sistemas dinâmicos a linearização é um método para **avaliar a estabilidade local** de um ponto de equilíbrio de um sistema de equações diferenciais não lineares ou sistemas dinãmicos discretos.
- A aproximação de modelos dinâmicos não lineares em modelos linearizados tem como objetivos fornecer **informações qualitativas** do comportamento do sistema **em torno de um ponto**.

Com relação ao ponto escolhido para a linearização temos:

- O ponto de linearização é o ponto com pequenas variações, ou seja, os sinais envolvidos tem que possuir pequena amplitude.
- No maior parte dos casos utiliza-se o **ponto de quilibrio**, também conhecido como ponto de operação ou estado estacionário.
