# <center>Instrumentação

### Instrumentação: Resumo

**Função:** Monitorar e controlar processos industriais.

**Locais dos Símbolos:**
- Fluxogramas de Processo e Engenharia
- Diagramas de Controle (P&I)
- Desenhos de Instrumentação
- Diagramas de Instalação e Ligação
- Plantas de Localização
- Diagramas Lógicos de Controle
- Listagem de Instrumentos
- Painéis Sinópticos e Semi-Gráficos
- Telas de Vídeo de Estações de Controle

**Aplicações:**
- Plataformas de Óleo e Gás
- Tratamento de Esgoto e Água
- Plantas de Energia

**Requisitos para Bom Desempenho:**
- Equipamentos e sistemas corretamente dimensionados
- Instrumentos de medição adequados
- Instalação e aferição apropriadas
- Controladores bem sintonizados
- Válvulas operando dentro das faixas projetadas

### Tipos de Sistemas Instrumentados

1. **Malha Aberta:** Controle sem ajuste baseado em feedback.
2. **Malha Fechada:** Controle ajustado com base em feedback.

### Componentes de Instrumentação

1. **Sensores:** Detectam variáveis do processo (ex.: temperatura, pressão).
2. **Transdutores:** Convertem variáveis em grandeza elétrica.
3. **Transmissores:** Combinam transdutor e condicionador de sinal para gerar um sinal padrão.

### Equipamentos e Instrumentos

1. **Equipamentos:** Bombas, tanques, motores, colunas.
2. **Instrumentos:** Indicadores, controladores, sensores, válvulas de controle.

### Normas

1. **Brasileiras:**
   - **ABNT NBR 8190:** Símbolos gráficos para instrumentação.

2. **Internacionais:**
   - **ISA 5.1:** Símbolos e identificação de instrumentação.
   - **ISA 5.2:** Diagramas Lógicos Binários.
   - **ISA 5.3:** Instrumentação de Display para Controle Distribuído.
   - **ISA 5.4:** Diagramas de Malha de Instrumentos.
   - **ISA 5.5:** Símbolos para Displays de Processo.

### Nomenclatura de Instrumentos

**Tagname:** Código alfanumérico para identificação e localização dos instrumentos.

- **Identificação Funcional:**
  - **1º Grupo de Letras:** Variável medida (ex.: P = Pressão).
  - **2º Grupo de Letras:** Função do instrumento (ex.: A = Alarme).

- **Exemplo de Tag:** PDAL-101-02 B
  - **P:** Pressão
  - **D:** Diferencial
  - **A:** Alarme
  - **L:** Baixo
  - **101:** Malha de controle
  - **02:** Instrumento específico
  - **B:** Redundância

- **Identificação da Malha:**
  - Número sequencial indicando a malha de controle.
  - Prefixos numéricos + número sequencial de 3 dígitos.

- **Dispositivos com Funções Múltiplas:**
  - Representar com vários símbolos.
  - Identificação funcional com máximo de 4 letras, todas maiúsculas.

#### **Indicadores**
- **PI:** Indicador de Pressão
  - **P:** Pressão
  - **I:** Função de Informação
  - Pode ser manômetro mecânico ou eletrônico.

- **TI:** Indicador de Temperatura
- **LI:** Indicador de Nível
- **SI:** Indicador de Velocidade
- **RI:** Indicador de Radioatividade
- **MI:** Indicador de Umidade
- **AI:** Indicador de Condutividade, pH, O₂, etc.
- **VI:** Indicador de Viscosidade

#### **Indicadores Controladores**
- **PIC:** Indicador Controlador de Pressão
  - **P:** Pressão
  - **IC:** Indicador e Controlador
  - **C:** Controle da malha

- **TIC:** Indicador Controlador de Temperatura
- **LIC:** Indicador Controlador de Nível
- **FIC:** Indicador Controlador de Vazão
- **JIC:** Indicador Controlador de Potência
- **SIC:** Indicador Controlador de Velocidade
- **BIC:** Indicador Controlador de Queima/Combustão

#### **Alarmes**
- **LAH:** Alarme de Nível Alto
  - **A:** Alarme
  - **H:** Nível Alto

- **TAH:** Alarme de Temperatura Alta
- **SAL:** Alarme de Baixa Velocidade
- **WAL:** Alarme de Peso Baixo

#### **Válvulas**
- **HV:** Válvula de Controle Manual
  - **V:** Válvula
  - **H:** Variável Inicial

- **LCV:** Válvula de Controle de Nível Auto-Operada
  - **C:** Auto-Operada

- **LV:** Válvula de Nível (Proporcional)

#### **Sistemas de Supervisão e Aquisição de Dados (SCADA)**
- **Função:** Monitorar e supervisionar variáveis e dispositivos de controle.
- **Relatórios Comuns:**
  - Relatório de Alarmes
  - Relatório de Acesso
  - Relatório de Variáveis

### Definições Adicionais

**4) O que são variáveis de processo?**
- Variáveis de processo são as quantidades ou condições que são medidas e controladas dentro de um sistema de processo, como temperatura, pressão, fluxo, nível, entre outros.

**5) Defina variável manipulada.**
- A variável manipulada é aquela que o sistema de controle ajusta para alcançar a condição desejada ou setpoint. Geralmente é a saída de um controlador que atua sobre o elemento final de controle, como a abertura de uma válvula ou a velocidade de uma bomba.

**6) Defina elemento final de controle.**
- Como mencionado, é o dispositivo que realiza uma ação direta sobre o processo com base no sinal de saída do controlador, como uma válvula, um atuador, ou outro mecanismo de ajuste direto no processo.
