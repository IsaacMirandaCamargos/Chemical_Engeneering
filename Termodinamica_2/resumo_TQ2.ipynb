<h1 align='center'>Relembrando Termodinâmica 1

<h2><u>Definições termodinamicas:</u>

A seguir iremos definir diversas grandezas termodinamicas para estabelecer relações entre a 1° lei e a 2° lei da termodinâmica.

- **Definições**:
    - Calor: $Q = TdS$
    - Trabalho: $W = -PdV$
    - Energia interna: $dU = Q + W$
    - Entalpia: $dH = dU + d(PV)$
    - Energia livre de Helmholtz: $dA = dU - d(TS)$
    - Energia livre de Gibbs: $dG = dH - d(TS)$

<h2><u>Relação 1° e 2° lei:</u>

A seguir, segue-se as definições termodinamicas expandidas aplicando a 1° e 2° lei da termodinamica:

- **Definições**:
    - <h3> dU = TdS - PdV</h3>
    - <h3> dH = TdS + VdP</h3>
    - <h3> dG = VdP - SdT</h3>
    - <h3> dA = - PdV - SdT</h3>

<h2><u>Grandezas termodinâmicas: Função de (P, T)</u>

Segue-se as definições das grandezas termodinâmicas em funçao de (P, T):

- **Definições**:
    - <h3>$ dS = \frac {Cp}{T}dT - \beta V dP  $</h3>
    - <h3>$ dU = V[KP - \beta T]dP - [Cp - PV \beta] dT  $</h3>
    - <h3>$ dH = Cp dT - V[1 - \beta T]dP  $</h3>
    - <h3>$ dG = V dP - SdT  $</h3>
    - <h3> dA = - PdV - SdT</h3>

<h1 align='center'>Propriedades Parciais Molares

<h2><u>Propriedade Parcial Molar:</u></h2>

Em termodinâmica, as propriedades parciais molares são uma forma de descrever como uma propriedade extensiva de uma mistura varia quando a quantidade de um dos componentes é alterada, mantendo a quantidade dos outros componentes, pressão e temperatura constantes. Elas são muito úteis para entender o comportamento de sistemas de múltiplas componentes, como soluções, misturas de gases, etc. 

- **Variação de propriedade de estado**:
    - #### $d(M) = \frac{d(M)}{dP}\bigg|_{T}dP + \frac{d(M)}{dT}\bigg|_{P}dT \quad \bullet \text{Sistema fechado} \therefore M = f(P, T)$
    - #### $d(nM) = \frac{d(nM)}{dP}\bigg|_{T, n}dP + \frac{d(nM)}{dT}\bigg|_{P, n}dT + \sum \frac{d(nM)}{dn_i}\bigg|_{T, P, n_j}dn_i \quad \bullet \text{Sistema fechado} \therefore M = f(P, T, n)$ 


- **Definição de Propriedade parcial molar**:

- ### $\boxed {\bar{M}_i = \frac{d(nM)}{dni} \bigg|_{P,T,n_j}}$

- **Propriedade**:

- ### $\boxed {M = \sum_i x_i \bar{M}_i}$


**Observação**: {M: Propriedade qualquer}

**Observação**: A propriedade do sistema é o somatório das propriedades parciais molares vezes suas frações molares.

- #### $\boxed {\mu_i = \bar{G}_i = \frac{d(nG)}{dni} \bigg|_{P,T,n_j}}$

Ademais, para o caso de um sistema contendo duas fases em equilibro chegamos que a energia livre de Gibbs parcial molar para a substância ( **i** ) é igual em ambas as fases

- **Sistema bifásico**:



- **Dedução**:

- #### $nG = (nG)^\alpha + (nG)^\beta$

- #### $d(nG) = d(nG)^\alpha + d(nG)^\beta$

- #### $d(nG) = (nV)dP - (nS)dT + \left( \sum \bar{G}_i dn_i \right)^\alpha + \left( \sum \bar{G}_i dn_i \right)^\beta \quad \bullet {T, P}_{const}$

- #### $\sum \bar{G}_i^\alpha dn_i^\alpha - \sum \bar{G}_i^\beta dn_i^\alpha = 0 \quad \bullet dn_i^\alpha = -dn_i^\beta$

- ### $\boxed {\bar{G}_i^\alpha = \bar{G}_i^\beta}$

<h1 align='center'>Modelos de Mistura

<h2><u>Modelos de mistura:</u></h2>

A seguir veremos como calcular as **propriedades parciais molares da substância (i) em mistura** em relação as **propriedades da substância (i) pura**.

<h2><u>Modelos de mistura (Gás ideal):</u></h2>

A seguir veremos como se comportam as **propriedades parciais molares da substância (i) em mistura** em relação as **propriedades da substância (i) pura**, para isso consideremos o gás como gás ideal.

- **Teorema de Gibbs**:
    - #### $\boxed{ \bar{M}^{gi}_{i} (T, P) = M^{gi}_{i} (T, p_i)}$



    

- **Volume parcial molar**: ($V^{gi} = \frac{RT}{P}$)

    
    - ### $\bar{V}^{gi}_i = \left(\frac{\partial(n \bar{V}^{gi})}{\partial n_i}\right)_{T,P,n_j} = \left(\frac{\partial(nRT/P)}{\partial n_i}\right)_{T,P,n_j} = \frac{RT}{P} \left(\frac{\partial n}{\partial n_i}\right)_{n_j} = \frac{RT}{P} \quad \boxed {\bar{V}^{gi}_i = \frac{RT}{P} }$

- **Pressão parcial molar**: ($yi = \frac {Pi}{P} = \frac {ni}{n}$)

    - ### $p_i = \frac{y_i RT}{V^{gi}_i} = y_i P \quad \boxed{p_i = y_i P}$

- **Outras propriedades**: (Pelo teorema de Gibbs; Entalpia independe da pressão; Energia interna independe da pressão)

    - #### $\boxed{\bar{H}^{gi}_{i} = H^{gi}_i} \quad \bullet \text{Entalpia}$
    
    - #### $\boxed{\bar{U}^{gi}_{i} = U^{gi}_i} \quad \bullet \text{Energia interna}$
    
    - #### $\boxed{\bar{G}^{gi}_{i} = G^{gi}_i + RT \ln y_i} \quad \boxed{\bar{G}^{gi}_{i} = \Gamma_i (T) + RT \ln(y_i P)} \quad \bullet \text{Energia livre de Gibbs}$
    
    - #### $\boxed{\bar{S}^{gi}_{i} = S^{gi}_i - R \ln y_i} \quad \bullet \text{Entropia}$

- **Dedução**:



**Observação**: $\tau (T)$ é uma constante em função da temperatura.

<h2><u>Modelos de mistura (Gás ideal):</u> Propriedade total da mistura</h2>

Agora, sabendo calcular as propriedades parciais molares de uma substância (i) em uma mistura, com base nas suas propriedades puras, iremos aprender como calcular as propriedades globais da mistura.

- **Propriedades globais da mistura (forma geral)**:

    - ### $\boxed{M^{gi} = \sum yi \bar M^{gi}_i}$

- **Então as propriedades da mistura ficam**:

- #### $\boxed{H^{gi} = \sum y_i H^{gi}_i} \quad \bullet \text{Entalpia}$
- #### $\boxed{U^{gi}_{i} = \sum y_i U^{gi}_i} \quad \bullet \text{Energia interna}$

- #### $\boxed{G^{gi}_{i} = \sum y_i G^{gi}_i + RT \sum y_i \ln y_i} \quad \boxed{G^{gi}_{i} = \sum y_i \Gamma_i (T) + RT \sum y_i \ln(y_i P)} \quad \bullet \text{Energia livre de Gibbs}$

- #### $\boxed{S^{gi}_{i} = \sum y_i S^{gi}_i - R \sum y_i \ln y_i} \quad \bullet \text{Entropia}$

<h2><u>Modelos de mistura (Gás ideal):</u> Delta de mistura</h2>

Sabemos como calcular a propriedade parcial molar ($\bar M_i$) de mistura a partir da sua propriedade pura. Também aprendemos como calcular a propriedade global da mistura e, por fim, iremos aprender como calcular a variação das propriedades no processo de mistura:

- **Delta de mistura (forma geral)**:
    - ### $\boxed{M^{gi} - \sum yi M^{gi}_i = \Delta Mistura }$ 

- **Então os delta de mistura ficam**:

- #### $\boxed{H^{gi} - \sum y_i H^{gi}_i = 0} \quad \bullet \text{Entalpia}$
- #### $\boxed{U^{gi} - \sum y_i U^{gi}_i = 0} \quad \bullet \text{Energia interna}$

- #### $\boxed{G^{gi} - \sum y_i G^{gi}_i = + RT \sum y_i \ln y_i} \quad \bullet \text{Energia livre de Gibbs}$

- #### $\boxed{S^{gi} - \sum y_i S^{gi}_i=  - R \sum y_i \ln y_i} \quad \bullet \text{Entropia}$

<h1 align='center'>Fugacidade

<h2><u>Fugacidade:</u></h2>

Até agora vimos como calcular diversas propriedades parciais molares e também como calcular as propriedades do sistema. Porem, o calculo da energia livre de Gibbs é de complicada representação e obtenção de seus valores. Para facilitar seu cálculo iremos definir uma nova grandeza **fugacidade**:

- **Fugacidade**:

    - ### $\boxed{f_i = P_{\text{real}}}$

- **Coeficiente de fugacidade**:

    - #### $\boxed{\phi_i = \frac{f_i}{P}}$

<h2><u>Fugacidade:</u> Energia livre de Gibbs.</h2>

A seguir veremos como calcular a energia livre de Gibbs atravéz da fugacidade e depois como calcular o **residuo** atrávez do coeficiente de fugacidade:

- **Relembrando**:

    - ### $M^R = M - M^{gi} \quad \bullet \text{Propriedades residuais}$
 
    - ### $G_i = \Gamma_i(T) + RT \ln P_{i} \quad \bullet \text{Energia livre de Gibbs}$

    
- **Calculo energia livre de Gibbs residual**:

    - #### $\boxed{G^R_i = RT \ln \phi_i}$

- **Dedução**:

- #### $G_i = \Gamma_i(T) + RT \ln P_{i, real} \quad \quad G^{gi}_i = \Gamma_i(T) + RT \ln P$
- #### $G_i - G^{gi}_i = RT \ln \frac{P_{i, real}}{P}$
- #### $G_i - G^{gi}_i = RT \ln \frac{f_i}{P}$
- #### $\boxed{G^R_i = RT \ln \phi_i}$

<h2><u>Fugacidade:</u> Energia livre de Gibbs Residual:

A seguir veremos como calcular a energia livre de Gibbs residual e consequentemente o coeficiente de fugacidade: (resumo completo em termo 1)

- **Pelas equações analiticas**:(Z: fator de compressibilidade **real**)

    - #### $\boxed{\frac{G^R}{RT} = \int_0^P (Z - 1)\frac{dP}{P}} \quad (T \text{ const.})$
    
- **Pela equação de estado virial**:($p$: Densidade molar $V^{-1}$)(Pressões moderadas)
    
    - #### $Z - 1 = \frac{BP}{RT} \quad \quad \quad Z - 1 = Bp + Cp^2$
    
    - #### $\boxed{\frac{G^R}{RT} = \frac{BP}{RT}} \quad \quad \boxed{\frac{G^R}{RT} = 2BP + \frac{3}{2} Cp^2 - \ln Z}$

- **Pelas equações cubicas**:

    - ###### **Raiz com caracteristica de líquido (1° chute: $Z=\beta $)** $\quad \quad \ \ $ **Raiz com caracteristica de vapor (1° chute: $Z=1$)**
    
    - #### $\boxed{Z = \beta + (Z + \varepsilon\beta)(Z + \sigma\beta)\left(\frac{1 + \beta - Z}{q\beta}\right)} \quad \quad \quad \quad \quad \quad \boxed{Z = 1 + \beta - q\beta \frac{Z - \beta}{(Z + \varepsilon\beta)(Z + \sigma\beta)}}$
    
    - ###### **Calculo $G^R$**:
    
    - ### $q = \frac{\Psi \alpha(T_r)}{\Omega T_r} \quad \quad \beta = \Omega\frac{P_r}{T_r} \quad \quad I = \frac{\beta}{Z + \varepsilon\beta} \quad \bullet \text{Para vdw} \quad \quad I = \frac{1}{\sigma - \varepsilon} \ln\left[\frac{Z + \sigma\beta}{Z + \varepsilon\beta}\right]  \quad \bullet \text{P/ as demais}$
    
    - #### $\boxed{\frac{G^R}{RT} = Z - 1 - \ln(Z - \beta) - q I}$

- **Correlações generalizadas de Lee e Kesler**:

    - #### $\boxed{\frac{G^R}{RT} = \ln \phi^0 + \omega \ln \phi^1} \quad \bullet \text{Tabelas E.13 a E.16}$

**Conclusão**: Calculando a energia livre residual calculamos o coeficiente de fugacidade.
