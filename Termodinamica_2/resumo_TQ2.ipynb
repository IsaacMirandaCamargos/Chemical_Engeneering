<h2><u>Definições termodinamicas:</u>

A seguir iremos definir diversas grandezas termodinamicas para estabelecer relações entre a 1° lei e a 2° lei da termodinâmica.

- **Definições**:
    - Calor: $Q = TdS$
    - Trabalho: $W = -PdV$
    - Energia interna: $dU = Q + W$
    - Entalpia: $dH = dU + d(PV)$
    - Energia livre de Helmholtz: $dA = dU - d(TS)$
    - Energia livre de Gibbs: $dG = dH - d(TS)$

<h2><u>Relação 1° e 2° lei:</u>

A seguir, segue-se as definições termodinamicas expandidas aplicando a 1° e 2° lei da termodinamica:

- **Definições**:
    - <h3> dU = TdS - PdV</h3>
    - <h3> dH = TdS + VdP</h3>
    - <h3> dA = - PdV - SdT</h3>
    - <h3> dG = VdP - SdT</h3>

Em termodinâmica, as propriedades parciais molares são uma forma de descrever como uma propriedade extensiva de uma mistura varia quando a quantidade de um dos componentes é alterada, mantendo a quantidade dos outros componentes, pressão e temperatura constantes. Elas são muito úteis para entender o comportamento de sistemas de múltiplas componentes, como soluções, misturas de gases, etc.
