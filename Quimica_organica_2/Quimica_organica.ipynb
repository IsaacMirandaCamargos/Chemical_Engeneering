<h1> # Ordem do alcool

Considerando o carbono **C** ligado a hidroxila **OH**, a  ordem do alcool é referente ao número de ligações que este carbono está se ligando:

- **Alcool primário**: O carbono ligado faz **uma ligação** com outro carbono.
- **Alcool secundário**: O carbono ligado faz **duas ligações** com outros carbonos da cadeia.
- **Alcool terciário**: O carbono ligado faz **três ligações** com outros carbonos da cadeia.

<h1> # Reações de SN em alcoóis: Haletos de hidrogênio

Os alcoóis assim como os haletos de alquila sofrem reações de substituição nucleofílica.

Desse modo, para que a reação **alcool** e **haletos de hidrogênio** ocorra será necessário meio ácido **(H<sup>+</sup>)** e aquecimento **(△)**, por exemplo:



Ademais, temos que as reações entre **alcoóis** e **haletos de alquila** ocorrem da seguinte forma:

- **Reações SN1**: Alcool secundário e terciário.
- **Reações SN2**: Alcool primário.

<h1> # Reação de SN1: Alcool e Haletos de hidrogênio

A reação de substituição nucleofilica unimolecular **SN1** em **alcoóis** e **haletos de hidrogênio** acontece apenas em **alcoóis secundários e terciários**, pelo seguinte mecanismo:

- Reação:
    - **Ativação do alcool**.
    - **Auto ionização** da H<sub>2</sub>O e **formação do carbocátion**.
    - Formação de dois produtos:
        - **Produto de substituição**: Ligação do haleto ao carbocátion.
        - **Produto de eliminação**:
            - **Majoritário**: Perda do hidrogênio pelo carbono de maior ordem.
            - **Outros**: Perda do hidrogênio pelos carbonos de ordens diferentes.
            


<h1> # Reação de SN2: Alcool e Haletos de hidrogênio

A reação de substituição nucleofilica bimolecular **SN2** em **alcoóis** e **haletos de hidrogênio** acontece apenas em **alcoóis primários**, pelo seguinte mecanismo:

- Reação:
    - **Ativação do alcool**.
    - **Ataque do haleto por trás** da H<sub>2</sub>O e **saída da H<sub>2</sub>O**.
    - Formação apenas do **produto de substituição**.
            


- Reação: (Com catalise por cloreto de zinco **ZnCl<sub>2</sub>**)
    - Ativação do alcool pelo **Zn<sup>2+</sup>**
    - **Ataque do haleto por trás** da OHZn<sup>1+</sup> e **saída do mesmo.



- Observação: O zinco não se liga ao oxigênio, ele apenas se **complexa**.
